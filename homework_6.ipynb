## Q1. penguins 데이터 프레임에서 각 종(species) 별로 flipper_length_mm의 histogram을 그리는 코드를 작성하세요. 단, matplotlib 라이브러리만 이용하세요.

```python
penguins.head()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(10, 10))
# your code here.
Adelie = penguins[penguins['species'] =='Adelie']
Chinstrap = penguins[penguins['species'] =='Chinstrap']
Gentoo = penguins[penguins['species'] =='Gentoo']

#plt.plot(penguins["species"], penguins["flipper_length_mm"])
#plt.xlabel('X-Label')
#plt.ylabel('Y-Label')
#plt.show()

#Adelie.head()
ax[0].hist(Adelie['flipper_length_mm'])
ax[0].set_xlabel("Adelie")
ax[1].hist(Chinstrap['flipper_length_mm'])
ax[1].set_xlabel("Chinstrap")
ax[2].hist(Gentoo['flipper_length_mm'])
ax[2].set_xlabel("Gentoo")

plt.show()
```

## Q2. penguins 데이터 프레임에서 펭귄의 종(species)을 X축으로 하고 body_mass_g의 평균을 보여주는 bar plot을 그리는 코드를 작성하세요. 이 때, 성별(sex)에 따라 bar가 stacked되도록 하세요. 단, matplotlib 라이브러리만 이용하세요.

```python
fig, ax = plt.subplots(figsize=(10,10))
# your code here.


Adelie_Female = Adelie[Adelie['sex'] =='Female']
Adelie_Male = Adelie[Adelie['sex'] =='Male']
Chinstrap_Female = Chinstrap[Chinstrap['sex'] =='Female']
Chinstrap_Male = Chinstrap[Chinstrap['sex'] =='Male']
Gentoo_Female = Gentoo[Gentoo['sex'] =='Female']
Gentoo_Male = Gentoo[Gentoo['sex'] =='Male']

Female = [Adelie_Female['body_mass_g'].mean(), Chinstrap_Female['body_mass_g'].mean(), Gentoo_Female['body_mass_g'].mean()]
Male = [Adelie_Male['body_mass_g'].mean(), Chinstrap_Male['body_mass_g'].mean(), Gentoo_Male['body_mass_g'].mean()]
species=["Adelie","Chinstrap","Gentoo"]

plt.bar(species, Female, color="blue",label="Female")
plt.bar(species, Male, color="yellow",bottom=Female,label="Male")

plt.legend(loc="lower left",bbox_to_anchor=(0.8,1.0))
plt.show()
```
